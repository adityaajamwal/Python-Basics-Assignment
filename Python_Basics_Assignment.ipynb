# Python Basics

**1.Explain the key features of Python that make it a popular choice for programming.**

Python is a highly popular programming language due to its versatility and beginner-friendly design. Its key features include:
- Readable and Easy to Learn: Python has a clean, English-like syntax that emphasizes readability, making it accessible for beginners.
- Interpreted Language: Python code is executed line-by-line. This makes debugging easier because the interpreter stops and reports errors immediately upon finding them.
- Dynamically Typed: You don't need to explicitly declare variable types. Python figures out the type at runtime (e.g., typing x = 10 automatically sets x as an integer).
- Extensive Standard Library: Python comes with a massive built-in library, meaning you don't have to write code from scratch for common tasks like regular expressions, file I/O, or web interaction.
- Platform Independent: Python code is cross-platform. A Python script written on Windows will run on macOS or Linux without needing modifications.

**3.Compare and contrast mutable and immutable objects in Python with examples**

Everything in Python is an object, and they fall into two categories regarding whether their internal state can be changed.
- Mutable Objects: These can be modified after they are created. If you change a mutable object, its memory address remains exactly the same.

      Examples: Lists, Dictionaries, Sets.

```python
#code
my_list = [1, 2, 3]
my_list[0] = 99      # Allowed. The list is now [99, 2, 3]
```

- Immutable Objects: These cannot be modified after creation. Attempting to alter an immutable object actually creates a completely new object in memory.

      Examples: Strings, Tuples, Integers.

```python
#code
my_string = "Hello"
# my_string[0] = "h"  # This would throw a TypeError! Strings cannot be changed.
```

**4.Discuss the different types of operators in Python and provide examples of how they are used.**

Operators are special symbols used to perform operations on variables and values.

- Arithmetic Operators: Used for mathematical operations.

      + (Addition), - (Subtraction), * (Multiplication), / (Division), % (Modulus/Remainder), // (Floor Division).

- Comparison (Relational) Operators: Compare two values and return a Boolean (True or False).

      == (Equal), != (Not equal), > (Greater than), < (Less than).

- Logical Operators: Used to combine conditional statements.

      and, or, not.

- Assignment Operators: Used to assign values to variables.

      =, +=, -=.

```python
#code example
x = 10 + 5                 # Arithmetic (+) and Assignment (=)
is_valid = (x > 10) and (x < 20)  # Comparison (>, <) and Logical (and)
```

**5.Explain the concept of type casting in Python with examples.**

Type casting is the process of converting a variable from one data type into another. This is highly useful when you need to perform operations that require matching data types.

-  Implicit Casting: Python automatically converts data types (e.g., adding an integer to a float automatically results in a float).

-  Explicit Casting: Manually converting data types using built-in functions like int(), float(), or str().

```python
#code example
# Converting a string to an integer for math
string_number = "50"
actual_number = int(string_number) + 10  # Results in 60

# Converting an integer to a string for concatenation
age = 25
message = "I am " + str(age) + " years old."
```

**6.How do conditional statements work in Python? Illustrate with examples.**

Conditional statements allow a program to make decisions. They evaluate a condition to determine if it is True or False, and then execute specific blocks of code based on the result.

- Python uses if, elif (else if), and else.

```python
#code example
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C or below")
```

**7.Describe the different types of loops in Python and their use cases with examples.**

Loops are used to execute a block of code repeatedly. Python primarily uses two types of loops:

- for Loop: Used to iterate over a sequence (like a list, tuple, dictionary, or string).

  Use Case: Best when you know exactly how many times you need to iterate, or when you are processing items in a specific collection.

```python
#code example
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

- while Loop: Executes a block of code repeatedly as long as a specified condition evaluates to True.

  Use Case: Best when the number of iterations is unknown beforehand and depends entirely on a dynamic condition.

```python
#code example
count = 0
while count < 3:
    print("Count is:", count)
    count += 1
```
